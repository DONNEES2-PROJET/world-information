```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données depuis le fichier CSV
data_frame = pd.read_csv('health-data.csv')


data_frame1 = data_frame[['Out of pocket health expenditure', 'Infant mortality']]
corr = data_frame1.corr(numeric_only=True)  # Spécifiez numeric_only comme True
sns.heatmap(corr, annot=True)
plt.show()

# Nettoyer et traiter les données si nécessaire

# Sélectionner uniquement les colonnes numériques pour le calcul de corrélation
numerical_columns = ['Out of pocket health expenditure', 'Infant mortality'] 
correlation = data_frame[numerical_columns].corr(numeric_only=True)
print("Matrice de corrélation :")
print(correlation)
```
