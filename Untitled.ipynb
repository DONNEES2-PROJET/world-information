```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les DataFrames
education_data = pd.read_csv('education-data.csv')
economics_data = pd.read_csv('economics-data.csv')

# Noms des colonnes
col_taux_scolarite = "Gross primary education enrollment (%)"
col_salaire = "Minimum wage"

# Sélectionner les colonnes pertinentes dans chaque DataFrame et copier explicitement
data_frame1 = education_data[[col_taux_scolarite]].copy()
data_frame2 = economics_data[[col_salaire]].copy()

# Remplacer les valeurs manquantes par zéro
data_frame1.fillna(0, inplace=True)
data_frame2.fillna(0, inplace=True)

# Convertir les colonnes en nombres décimaux
data_frame1[col_taux_scolarite] = pd.to_numeric(data_frame1[col_taux_scolarite], errors='coerce')
data_frame2[col_salaire] = pd.to_numeric(data_frame2[col_salaire], errors='coerce')


data_frame1[col_taux_scolarite] = (data_frame1[col_taux_scolarite] - data_frame1[col_taux_scolarite].min()) / (data_frame1[col_taux_scolarite].max() - data_frame1[col_taux_scolarite].min())
data_frame2[col_salaire] = (data_frame2[col_salaire] - data_frame2[col_salaire].min()) / (data_frame2[col_salaire].max() - data_frame2[col_salaire].min())

# Calculer la corrélation entre les colonnes normalisées
corr = data_frame1.corrwith(data_frame2)

# Afficher la heatmap de la corrélation
sns.heatmap(pd.DataFrame(corr), annot=True)
plt.title("Corrélation entre Taux de scolarité et Salaire minimum")
plt.show()
```
